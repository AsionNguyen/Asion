* `cinemasci` est un ensemble d'outils `python` pour lire, créer et visualiser des objets du type `Cinema database (.cdb)`

    + ...

## `Cinema_view` vs `Paraview_glance`

Types| Taille |
:-----|:------
.csv <br>([donnée originale][view]) |12Mo    |
.cdb <br>(Cinema_view)|7.4Mo : $\phi = 6, \theta =6$, images de base resolution <br> 9.7Mo : $\phi = 6, \theta =6$, images de normale resolution <br> 19.4Mo: $\phi = 6, \theta =6$, images de haute resolution <br> [116Mo][link]: $\phi = 36, \theta =6$, images de haute resolution |
.vtkjs <br>(Paraview_glance)|8.7Mo|

[view]: https://raw.githubusercontent.com/AsionNguyen/paraviewglance/main/3.000_0.csv "point"
[link]: https://www.youtube.com/watch?v=9prBGsJLul4 "cine"

* Avantage de `cinema_view` est que la taille d'objet dépend du choix de résolution (en $\phi, \theta$ et d'image) pas de donnée originale. Donc si la donnée .csv est très grosse, son .cbd resterait petite. Par ailleurs, `cinema_view` profite de `slide bar` pour la visualisation 360.
* Avantage de `paraview_glance` est l'interactivité diverse sur la donnée originale. Le rapport entre les tailles de ces deux structures n'a pas encore été étudié.
